```python
data = data.astype(int)
data['Class'] = data['Class'].replace({2: 0, 4: 1})
data['Class'].unique()
```

```python
data.columns = ["clump_thickness", "uniformity_of_cell_size", "uniformity_of_cell_shape", "marginal_adhesion", "single_epithilial_cell_size", "bare_nuclei", "bland_chromatin", "normal_nucleoli", "mitosis", "Class"]
data.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop("Class", axis=1), data["Class"], test_size=0.2, random_state=42)
X_train = (X_train - np.min(X_train))/(np.max(X_train) - np.min(X_train)).values
X_test = (X_test - np.min(X_test))/(np.max(X_test) - np.min(X_test)).values
```

```python
for column in data.columns[:-1]:
    formula = f"Class ~ {column}"
    model = sm.GLM.from_formula(formula, family=sm.families.Binomial(), data=data)
    res = model.fit()
    print(f"Simple Logistic Regression for {column}:")
    print(res.summary())
    stats=pd.DataFrame(np.exp(res.params),columns=['OddsRatio'])
    stats['P>|z|']=res.pvalues
    stats[['2.5%','97.5%']]=np.exp(res.conf_int())
    print(stats)
    print("")
    print("##########################################################################")
    print("")
```

```python
import itertools
best_subset = None
best_aic = float('inf')

cols = data.drop('Class',axis=1).columns

# Create feature subsets of different lengths
for subset_length in range(2, len(cols) + 1):

    # Generate all possible combinations of columns with the current subset length
    column_combinations = itertools.combinations(cols, subset_length)

    for columns in list(column_combinations):
        formula = f"Class ~ {' + '.join(columns)}"

        model = sm.GLM.from_formula(formula, family=sm.families.Binomial(), data=data)
        res = model.fit()
        current_aic = res.aic

        # Check if the current AIC is better than the best found so far
        if current_aic < best_aic and all(res.pvalues[1:] < 0.05):
            best_aic = current_aic
            best_subset = columns

# Now, we have the best subset of features with the lowest AIC
print("Best subset with the lowest AIC:", best_subset)
print("Lowest AIC value:", best_aic)
```

Multiple Logistic Regression results-

It usually is best to select the model with the smaller AIC. AIC (Akaike Information Criterion) is used to compare the performance of models on the same dataset.  After generating all possible different combinations of variable , decided the best combination
of independent variables (displayed in above cell) for my multiple logistic regression model using the lowest AIC and pvalue < 0.05.
